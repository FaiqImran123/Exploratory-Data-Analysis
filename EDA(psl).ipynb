## Data Wrangling

### Summary About Dataset
Dataset contains almost 50k records and 16 columns. It contains record each and every ball during 2016 - 2022 PSL

### Column Description:
- psl_year: The year of the match, ranging from 2016 to 2022.
- match_number: A unique identifier for each match.
- team_1: The name or identifier of the first team participating in the match.
- team_2: The name or identifier of the second team participating in the match.
- inning: Indicates the inning number (1st inning, 2nd inning, etc.).
- over: The over number in which the ball was bowled.
- ball: The ball number within the over.
- runs: The number of runs scored off the ball.
- wicket: Nature of eicket (0: "Caught", 1: "Bowled", 3:"lbw", 4:"run_out", 5: "Stumped", 6:"Hit Wicket", 8:"obstructfield")
- total_runs: The total runs scored up to this ball in the inning
- wickets: The total number of wickets fallen up to this ball in the inning.
- is_four: Indicates whether the runs scored off this ball resulted in a boundary
- is_six: Indicates whether the runs scored off this ball resulted in a six
- is_wicket: Indicates whether this ball resulted in a wicket
- wicket_text: Additional text describing the nature of the wicket, such as "caught", "bowled", "run out", etc.
- result: The result of the match

### Data Cleaning

```python
# 1: "Caught", 2: "Bowled", 3:"lbw", 4:"run_out", 5: "Stumped", 6:"Hit Wicket", 8:"obstructfield"
def apply_category(wc):
    if wc==1:
        return "caught"
    elif wc==2:
        return "bowled"
    elif wc==3:
        return "lbw"
    elif wc==4:
        return "run out"
    elif wc==5:
        return "stumped"
    elif wc==6:
        return "hit wicket"
    elif wc==8:
        return "obstruct field"
    
psl_new["wicket_text"] =psl["wicket"].apply(apply_category)
```

### Team1 and Team2

```python
# count number of matches played by each team
def new(rec):
    return rec["team_1"].unique()[0]
series =psl_new.groupby(["psl_year", "match_number"]).apply(func =new)
series_1 = series.value_counts()
series_1
```

```python
def new_1(rec):
    return rec["team_2"].unique()[0]
series =psl_new.groupby(["psl_year", "match_number"]).apply(func =new_1)
series_2 =series.value_counts()
series_2
```

```python
series = series_1 + series_2
series =series.sort_values(ascending =False)
```

```python
# number of matches played by each team in entire psl 
series.plot(kind ="bar", title="Number of Matches Played By Each Team in Enitre PSL", xlabel ="Team", ylabel="Number of Matches")

```
